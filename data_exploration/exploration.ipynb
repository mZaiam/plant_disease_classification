```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from matplotlib.patches import Patch
```

```python
data_plantvillage = {
    'Ma': {'Total': 3172, 'Saudvel': 1645},
    'Mirtilo': {'Total': 1502, 'Saudvel': 1502},
    'Cereja': {'Total': 1906, 'Saudvel': 854},
    'Milho': {'Total': 3852, 'Saudvel': 1162},
    'Uva': {'Total': 4063, 'Saudvel': 423},
    'Laranja': {'Total': 5507, 'Saudvel': 0},
    'Pssego': {'Total': 2657, 'Saudvel': 360},
    'Pimento': {'Total': 2475, 'Saudvel': 1478},
    'Batata': {'Total': 2152, 'Saudvel': 152},
    'Framboesa': {'Total': 371, 'Saudvel': 371},
    'Soja': {'Total': 5090, 'Saudvel': 5090},
    'Morango': {'Total': 1565, 'Saudvel': 456},
    'Abbora': {'Total': 1835, 'Saudvel': 0},
    'Tomate': {'Total': 18162, 'Saudvel': 1592},
}

df = pd.DataFrame.from_dict(data_plantvillage, orient='index')
df['Doente'] = df['Total'] - df['Saudvel']
df_sorted = df.sort_values(by='Total', ascending=False)

plt.figure(figsize=(10, 5))

plt.bar(df_sorted.index, df_sorted['Doente'], color='blue', label='Fitopatologia', edgecolor='k')
plt.bar(df_sorted.index, df_sorted['Saudvel'], bottom=df_sorted['Doente'], color='green', label='Saudvel', edgecolor='k')

plt.ylabel('Nmero de Imagens')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
total = 0
saudavel = 0

for i in list(data_plantvillage.keys()):
    total += data_plantvillage[i]['Total']
    saudavel += data_plantvillage[i]['Saudvel']

print(f'Razo de classes saudveis: {((saudavel / total)*100):.2f}%')
```

## Data from FAO

```python
df_usa = pd.read_csv("data_eua.csv")
df_usa_plot = df_usa.sort_values('Value', ascending=False).head(10)
labels_usa_plot = [
    'Milho',
    'Soja',
    'Trigo',
    'Beterraba',
    'Cana de Acar',
    'Batata',
    'Tomate',
    'Arroz',
    'Semente \nde Algodo', 
    'Sorgo'
]

colors = [
    'green',
    'green',
    'red',
    'red',
    'red',
    'green',
    'green',
    'red',
    'red',
    'red'
]

plt.figure(figsize=(12, 5))

plt.bar(
    labels_usa_plot,
    df_usa_plot['Value'].values / 10e6,
    edgecolor='k',
    color=colors
)
plt.ylabel('Produo em Massa (Mt)')

legend_handles = [
    Patch(facecolor='green', edgecolor='k', label='Presente na base PlantVllage'),
    Patch(facecolor='red', edgecolor='k', label='Ausente na base PlantVllage') 
]

plt.legend(handles=legend_handles)
plt.xticks(rotation=45, ha='right')
plt.show()
```
