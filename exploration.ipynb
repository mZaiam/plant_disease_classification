```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_digipathos = {
    'Pimenta do Reino': {'Total': 15, 'Saudável': 3},
    'Feijão': {'Total': 129, 'Saudável': 2},
    'Cajueiro': {'Total': 131, 'Saudável': 0},
    'Mandioca': {'Total': 134, 'Saudável': 14},
    'Citros': {'Total': 135, 'Saudável': 8},
    'Coqueiro': {'Total': 308, 'Saudável': 0},
    'Café': {'Total': 190, 'Saudável': 4},
    'Milho': {'Total': 181, 'Saudável': 0},
    'Algodão': {'Total': 124, 'Saudável': 0},
    'Cupuaçu': {'Total': 5, 'Saudável': 0},
    'Videira': {'Total': 83, 'Saudável': 2},
    'Couve': {'Total': 8, 'Saudável': 1},
    'Melão': {'Total': 9, 'Saudável': 0},
    'Palma de Óleo': {'Total': 16, 'Saudável': 0},
    'Mamoeiro': {'Total': 26, 'Saudável': 0},
    'Maracujá': {'Total': 142, 'Saudável': 6},
    'Abacaxi': {'Total': 19, 'Saudável': 0},
    'Arroz': {'Total': 6, 'Saudável': 0},
    'Soja': {'Total': 491, 'Saudável': 9},
    'Cana de Açúcar': {'Total': 123, 'Saudável': 5},
    'Trigo': {'Total': 51, 'Saudável': 7}
}

df = pd.DataFrame.from_dict(data_digipathos, orient='index')
df['Doente'] = df['Total'] - df['Saudável']
df_sorted = df.sort_values(by='Total', ascending=False)

plt.figure(figsize=(10, 5))

plt.bar(df_sorted.index, df_sorted['Doente'], color='blue', label='Fitopatologia', edgecolor='k')
plt.bar(df_sorted.index, df_sorted['Saudável'], bottom=df_sorted['Doente'], color='green', label='Saudável', edgecolor='k')

plt.ylabel('Número de Imagens')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
total = 0
saudavel = 0

for i in list(data_digipathos.keys()):
    total += data_digipathos[i]['Total']
    saudavel += data_digipathos[i]['Saudável']

print(f'Razão de classes saudáveis: {((saudavel / total)*100):.2f}%')
```
