```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_digipathos = {
    'Pimenta do Reino': {'Total': 15, 'Saudvel': 3},
    'Feijo': {'Total': 129, 'Saudvel': 2},
    'Cajueiro': {'Total': 131, 'Saudvel': 0},
    'Mandioca': {'Total': 134, 'Saudvel': 14},
    'Citros': {'Total': 135, 'Saudvel': 8},
    'Coqueiro': {'Total': 308, 'Saudvel': 0},
    'Caf': {'Total': 190, 'Saudvel': 4},
    'Milho': {'Total': 181, 'Saudvel': 0},
    'Algodo': {'Total': 124, 'Saudvel': 0},
    'Cupuau': {'Total': 5, 'Saudvel': 0},
    'Videira': {'Total': 83, 'Saudvel': 2},
    'Couve': {'Total': 8, 'Saudvel': 1},
    'Melo': {'Total': 9, 'Saudvel': 0},
    'Palma de leo': {'Total': 16, 'Saudvel': 0},
    'Mamoeiro': {'Total': 26, 'Saudvel': 0},
    'Maracuj': {'Total': 142, 'Saudvel': 6},
    'Abacaxi': {'Total': 19, 'Saudvel': 0},
    'Arroz': {'Total': 6, 'Saudvel': 0},
    'Soja': {'Total': 491, 'Saudvel': 9},
    'Cana de Acar': {'Total': 123, 'Saudvel': 5},
    'Trigo': {'Total': 51, 'Saudvel': 7}
}

df = pd.DataFrame.from_dict(data_digipathos, orient='index')
df['Doente'] = df['Total'] - df['Saudvel']
df_sorted = df.sort_values(by='Total', ascending=False)

plt.figure(figsize=(10, 5))

plt.bar(df_sorted.index, df_sorted['Doente'], color='blue', label='Fitopatologia', edgecolor='k')
plt.bar(df_sorted.index, df_sorted['Saudvel'], bottom=df_sorted['Doente'], color='green', label='Saudvel', edgecolor='k')

plt.ylabel('Nmero de Imagens')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
total = 0
saudavel = 0

for i in list(data_digipathos.keys()):
    total += data_digipathos[i]['Total']
    saudavel += data_digipathos[i]['Saudvel']

print(f'Razo de classes saudveis: {((saudavel / total)*100):.2f}%')
```
